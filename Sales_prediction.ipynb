Sales prediction involves forecasting the amount of product customers will purchase, considering various factors like expenditure, advertising, etc

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing dataset
df=pd.read_csv("/content/Advertising.csv")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
#On the basic observation, average expenses on Tv are high and lowest on radio. lets visualize
sns.pairplot(df,x_vars=['TV','Radio','Newspaper'],y_vars='Sales',kind='scatter')
plt.show()
```

When advertising cost increase in TV Ads , the sales are also increasing . This is not clearly observed in case of Radio and newspaper

```python
df['TV'].plot.hist(bins=10)
```

```python
df['Radio'].plot.hist(bins=10,color='teal')
```

```python
df['Newspaper'].plot.hist(bins=10,color='grey')
```

```python
sns.heatmap(df.corr(),annot=True)
plt.show()
```

**Based on the observations of histogram and heatmap , sales of TV are having high impact**


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[['TV']],df[['Sales']],test_size=0.4,random_state=0)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model =LinearRegression()
model.fit(x_train,y_train)
```

```python
res=model.predict(x_test)
print(res)
```

```python
plt.plot(res)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,7.19448879+0.04613525*x_test,'r')
plt.show()
```

Concluding with saying that above mentioned solution was successful to predict sales using advertising platform datasets
